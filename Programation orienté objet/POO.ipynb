```python
import tutor
def positif(pile):
    pile1 = list(pile)
    pile2 = []
    while pile1 != []:
        x = pile1.pop()
        if x >= 0:
            pile2.append(x)
    while pile2 != []:
        x = pile2.pop()
        pile1.append(x)
    return pile1

positif([-1,0,5,-3,4,-6,10,9,-8])

tutor.tutor()

```

```python
class Pile:
    """
    Cette classe implémente en POO le TAD pile
    """
    def __init__(self):
        self.p=[]
    def tester(self):
        return self.p==[]
    def empiler(self,x):
        self.p.append(x)
    def depiler(self):
        return self.p.pop()
    def __str__(self):
        return str(self.p)
    
    
```
