```python
def indices_maxi(tab):
    '''
    >>> indices_maxi([1, 5, 6, 9, 1, 2, 3, 7, 9, 8])
    (9, [3, 8])
    >>> indices_maxi([7])
    (7, [0])
    '''
    imax=tab[0]
    indices=[]
    for i in range(len(tab)):
        if imax<tab[i]:
            imax=tab[i]
    for j in range(len(tab)):
        if tab[j]==imax:
            indices.append(j)
    return (imax,indices)
```

```python
indices_maxi([1, 5, 6, 9, 1, 2, 3, 7, 9, 8])
```

```python
indices_maxi([7])
```

```python
def renverse(pile):
    '''renvoie une pile contenant les mêmes éléments que pile,
    mais dans l'ordre inverse.
    Cette fonction détruit pile.
    >>> renverse([1, 2, 3, 4, 5])
    [5, 4, 3, 2, 1]'''
    pile_inverse = []
    while pile != []:
        pile_inverse.append(pile.pop())
    return pile_inverse
def positifs(pile):
    '''renvoie une pile contenant les éléments positifs de pile,
    dans le même ordre. Cette fonction détruit pile.
    >>> positifs([-1, 0, 5, -3, 4, -6, 10, 9, -8])
    [0, 5, 4, 10, 9]
    >>> positifs([-2])
    []
    '''
    pile_positifs = []
    while pile != []:
        u= pile.pop()
        if u >= 0:
            pile_positifs.append(u)
    return renverse(pile_positifs)
```

```python
renverse([1, 2, 3, 4, 5])
```

```python
positifs([-1, 0, 5, -3, 4, -6, 10, 9, -8])
```

```python
positifs([-2])
```
