```python
def moyenne(tab):
    """
    moyenne([(15.0,2),(9.0,1),(12.0,3)])
    12.5
    """
    num = 0
    denom = 0
    for i in range(len(tab)):
        num += tab[i][0] * tab[i][1]
        denom += tab[i][1]
    return num/denom
```

```python
moyenne([(15.0,2),(9.0,1),(12.0,3)])
```

```python
def ligne_suivante(ligne): 
    '''Renvoie la ligne suivant ligne du triangle de Pascal
    >>> ligne_suivante([1, 3, 3, 1]) 
    [1, 4, 6, 4, 1]
    ''' 
    ligne_suiv = [1] 
    for i in range(1,len(ligne)): 
        ligne_suiv.append(ligne[i-1]+ligne[i]) 
    ligne_suiv.append(1) 
    return ligne_suiv 
    
def pascal(n): 
    '''Renvoie le triangle de Pascal de hauteur n
    >>> pascal(2) 
    [[1], [1, 1], [1, 2, 1]] 
    >>> pascal(3) 
    [[1], [1, 1], [1, 2, 1], [1, 3, 3, 1]]
    ''' 
    triangle = [ [1] ] 
    for k in range(n): 
        ligne_k = ligne_suivante(triangle[k]) 
        triangle.append(ligne_k) 
    return triangle
```

```python
ligne_suivante([1, 3, 3, 1]) 
```
