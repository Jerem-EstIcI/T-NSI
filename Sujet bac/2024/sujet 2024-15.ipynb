```python
def moyenne(tab):
    '''
    >>> moyenne([1.0])
    1.0
    >>> moyenne([1.0, 2.0, 4.0])
    2.3333333333333335

    '''
    somme=0
    for i in range(len(tab)):
        somme+=tab[i]
    return somme/len(tab)
```

```python
moyenne([1.0])
```

```python
moyenne([1.0, 2.0, 4.0])
```

```python
def binaire(a):
    '''convertit un nombre entier a en sa representation
    binaire sous forme de chaine de caractÃ¨res.
    >>> binaire(83)
    '1010011'
    >>> binaire(6)
    '110'
    >>> binaire(127)
    '1111111'
    >>> binaire(0)
    '0'
    '''
    if a == 0:
        return '0'
    bin_a = str()
    while a>0:
        bin_a = str(a%2)  + bin_a
        a = a//2
    return bin_a
```

```python
binaire(83)
```

```python
binaire(6)
```

```python
binaire(127)
```

```python
binaire(0)
```
