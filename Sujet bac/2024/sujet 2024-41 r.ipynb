```python
class Noeud:
    def __init__(self, etiquette, gauche, droit):
        '''
        >>>a = Noeud(1, Noeud(4, None, None),Noeud(0, None,Noeud(7, None, None)))
        '''
        self.v = etiquette
        self.gauche = gauche
        self.droit = droit
def hauteur(a):
    '''
    >>> hauteur(a)
    2
    >>> hauteur(None)
    -1
    >>> hauteur(Noeud(1, None, None))
    0
    '''
    if a is None:
        return -1
    else:
        return 1 + max(hauteur(a.gauche), hauteur(a.droit))
def taille(a):
    '''
    >>> taille(a)
    4
    >>> taille(None)
    0
    >>> taille(Noeud(1, None, None))
    1
    '''
    if a==None:
        return 0
    else:
        return 1 + taille(a.gauche)+taille(a.droit)
```

```python
def ajoute(indice, element, tab):
    '''Renvoie un nouveau tableau obtenu en insérant
    element à l'indice indice dans le tableau tab.
    >>> ajoute(1, 4, [7, 8, 9])
    [7, 4, 8, 9]
    >>> ajoute(3, 4, [7, 8, 9])
    [7, 8, 9, 4]
    >>> ajoute(0, 4, [7, 8, 9])
    [4, 7, 8, 9]
    '''
    nbre_elts = len(tab)
    tab_ins = [0] * (nbre_elts + 1)
    for i in range(indice):
        tab_ins[i] = tab[i]
    tab_ins[indice] = element
    for i in range(indice + 1, nbre_elts + 1):
        tab_ins[i] = tab[i-1]
    return tab_ins
```

```python
ajoute(1, 4, [7, 8, 9])
```

```python
ajoute(3, 4, [7, 8, 9])
```

```python
ajoute(0, 4, [7, 8, 9])
```
