```python
def recherche(tab,occ):
    """
    >>> recherche([5, 3],1) # renvoie None 
    >>> recherche([2,4],2) 
    0 
    >>> recherche([2,3,5,2,4],2) 
    3
    """
    iocc=None
    for i in range(len(tab)):
        if tab[i]==occ:
            iocc=i
    return iocc
```

```python
recherche([5, 3],1)
```

```python
recherche([2,4],2)
```

```python
recherche([2,3,5,2,4],2)
```

```python
def distance_carre(point1, point2): 
    
    """ 
    Calcule et renvoie la distance au carre entre deux points.
    >>> distance_carre((1, 0), (5, 3)) 
    25 
    >>> distance_carre((1, 0), (0, 1)) 
    2
    """ 
    return (point1[0]-point2[0])**2 + (point1[1]-point2[1])**2 

def point_le_plus_proche(depart, tab): 
    """ Renvoie les coordonnées du premier point du tableau tab 
    se trouvant à la plus courte distance du point depart.
    >>> point_le_plus_proche((0, 0), [(7, 9), (2, 5), (5, 2)]) 
    (2, 5) 
    >>> point_le_plus_proche((5, 2), [(7, 9), (2, 5), (5, 2)]) 
    (5, 2)""" 
    min_point = tab[0] 
    min_dist = distance_carre(min_point, depart) 
    for i in range(1, len(tab)): 
        if distance_carre(tab[i], depart) < min_dist: 
            min_point = tab[i] 
            min_dist = distance_carre(tab[i], depart) 
    return min_point
```

```python
distance_carre((1, 0), (5, 3)) 
```

```python
distance_carre((1, 0), (0, 1)) 
```

```python
point_le_plus_proche((0, 0), [(7, 9), (2, 5), (5, 2)])
```

```python
point_le_plus_proche((5, 2), [(7, 9), (2, 5), (5, 2)]) 
```
