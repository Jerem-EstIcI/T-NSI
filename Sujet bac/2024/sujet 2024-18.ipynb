```python
def multiplication(n1,n2):
    '''
    >>> multiplication(3, 5)
    15
    >>> multiplication(-4, -8)
    32
    >>> multiplication(-2, 6)
    -12
    >>> multiplication(-2, 0)
    0
    '''
    somme=0
    signe=1
    if n2==0 or n1==0:
        return 0
    if n2<0 and n1 <0:
        n2=-n2
        n1=-n1
    if n2<0:
        n2=-n2
        signe=-1
    for i in range(n2):
        somme+=n1
    if signe == 1:
        return somme
    else:
        return -somme
```

```python
def chercher(tab, x, i, j):
    '''Renvoie l'indice de x dans tab, si x est dans tab,
    None sinon.
    On suppose que tab est triÃ© dans l'ordre croissant.
    >>> chercher([1, 5, 6, 6, 9, 12], 7, 0, 10)
    >>> chercher([1, 5, 6, 6, 9, 12], 7, 0, 5)
    >>> chercher([1, 5, 6, 6, 9, 12], 9, 0, 5)
    4
    >>> chercher([1, 5, 6, 6, 9, 12], 6, 0, 5)
    2
    '''
    if i > j:
        return None
    m = (i + j) // 2
    if tab[m] < x:
        return chercher(tab, x, m , j)
    elif tab[m] > x:
        return chercher(tab, x, j , m)
    else:
        return m
```
