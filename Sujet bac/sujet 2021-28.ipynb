```python
a={'F':['B','G'], 'B':['A','D'], 'A':['',''], 'D':['C','E'], 'C':['',''], 'E':['',''], 'G':['','I'], 'I':['','H'], 'H':['','']}
```

```python
def taille(a,lettre):
    '''
    >>>taille(a,'F')
    9
    '''
    k=a[lettre]
    md=k[1]
    mg=k[0]
    if mg=="" and md=="":
        return 1
    elif mg=="" and md!="":
        return 1 + taille(a,md)
    elif mg!="" and md=="":
        return 1 + taille(a,mg)
    else:
        return 1 + taille(a,mg) + taille(a,md)
    
```

```python
taille(a,'F')
```

```python
def tri_interatif(tab):
    '''
    >>>tri_interatif([41, 55, 21, 18, 12, 6, 25])
    [6, 18, 12, 21, 25, 41, 55]
    '''
    for k in range(len(tab) -1,0,-1):
        imax = k
        for i in range(0,k):
            if tab[i] > tab[imax]:
                imax = i
        if tab[imax] > tab[k]:
            tab[k], tab[imax] = tab[imax], tab[k]
    return tab
```

```python
tri_interatif([41,55,21,18,12,6,25])
```
