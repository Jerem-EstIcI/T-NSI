```python
def verifie(tab):
    '''
    >>> verifie([0, 5, 8, 8, 9])
    True
    >>> verifie([8, 12, 4])
    False
    >>> verifie([-1, 4])
    True
    >>> verifie([5])
    True
    '''
    for i in range(len(tab)-1):
        if tab[i]>tab[i+1]:
            return False
    return True
```

```python
verifie([0, 5, 8, 8, 9])
```

```python
verifie([8, 12, 4])
```

```python
urne = ['A', 'A', 'A','B', 'C', 'B', 'C','B', 'C', 'B']

def depouille(urne):
    resultat = dict()
    for bulletin in urne:
        if bulletin in resultat:
            resultat[bulletin] += 1
        else:
            resultat[bulletin] = 1
    return resultat

def vainqueur(election):
    vainqueur = ''
    nmax = 0
    for candidat in election:
        if election[candidat] > nmax:
            nmax = election[candidat]
            vainqueur = candidat
    liste_finale = [nom for nom in election if election[nom] == nmax]
    return liste_finale
```

```python
depouille(['A', 'A', 'A','B', 'C', 'B', 'C','B', 'C', 'B'])
```

```python
vainqueur(depouille(['A', 'A', 'A','B', 'C', 'B', 'C','B', 'C', 'B']))
```
