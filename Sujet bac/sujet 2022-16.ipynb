```python
def maxi(tab):
    '''
    >>>maxi([1,5,6,9,6,5,7,7,8,9,7,6])
    (9,3)
    '''
    nb = tab[0]  
    nbi = 0    

    for i in range(1, len(tab)):
        if tab[i] > nb:
            nb = tab[i]
            nbi = i

    return nb, nbi
```

```python
maxi([1,5,6,9,6,5,7,7,8,9,7,6])
```

```python
def positif(T):
    '''
    >>> positif([-1, 0, 5, -3, 4, -6, 10, 9, -8])
    T = [-1, 0, 5, -3, 4, -6, 10, 9, -8]
    [0, 5, 4, 10, 9]
    '''
    T2 = list(T)  # copie de la liste T
    T3 = []

    while T2 != []:
        x = T2.pop()  # supprime le dernier élément de T2
        if x >= 0:
            T3.append(x)  # ajoute x à la liste T3 si positif

    T2 = []
    while T3 != []:
        x = T3.pop()  # supprime le dernier élément de T3
        T2.append(x)  # ajouter x à la liste T2

    print('T =', T)
    return T2
```

```python
positif([-1, 0, 5, -3, 4, -6, 10, 9, -8])
```
