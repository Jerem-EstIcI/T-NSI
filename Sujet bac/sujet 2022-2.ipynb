```python
def moyenne(t):
    k=0
    l=0
    i=0
    while i<len(t):
        k+=t[i][0]*t[i][1]
        l+=t[i][1]
        i+=1
    return k/l

moyenne([(15,2),(9,1),(12,3)])
```

```python
def pascal(n):
    C= [[1]]
    for k in range(1,n):
        Ck = [1]
        for i in range(1,k):
            Ck.append(C[k-1][i-1]+C[k-1][i] )
        Ck.append(1)
        C.append(Ck)
    return C
```

```python
pascal(4)
```

```python
pascal(5)
```
