```python
def effectif_notes(tab):
    '''
    >>> notes_eval = [2, 0, 5, 9, 6, 9, 10, 5, 7, 9, 9, 5, 0, 9, 6, 5, 4] 
    >>> eff = effectif_notes(notes_eval) 
    >>> eff 
    [2, 0, 1, 0, 1, 4, 2, 1, 0, 5, 1]
    '''
    effectif=[]
    nb=0
    n=0
    while nb!=11:
        for i in range(len(tab)):
            if tab[i]==nb:
                n+=1
        nb+=1
        effectif.append(n)
        n=0
    return effectif
```

```python
notes_eval = [2, 0, 5, 9, 6, 9, 10, 5, 7, 9, 9, 5, 0, 9, 6, 5, 4] 
eff = effectif_notes(notes_eval) 
eff 
```

```python
def notes_triees(tab):
    '''
    >>> notes_triees([2, 0, 1, 0, 1, 4, 2, 1, 0, 5, 1]) 
    [0, 0, 2, 4, 5, 5, 5, 5, 6, 6, 7, 9, 9, 9, 9, 9, 10]
    '''
    note=[]
    for i in range(len(tab)):
        for nb in range(tab[i]):
            note.append(i)
    return note
```

```python
def dec_to_bin(nb_dec):
    '''
    >>> dec_to_bin(25) 
    '11001'
    '''
    q, r = nb_dec // 2, nb_dec % 2 
    if q == 0: 
        return str(r) 
    else: 
        return dec_to_bin(q) + str(r) 
    
def bin_to_dec(nb_bin):
    '''
    >>> bin_to_dec('101010') 
    42
    '''
    if len(nb_bin) == 1:
        if nb_bin == '0':
            return 0
        else:
            return 1
    else:
        if nb_bin[-1] == '0':
            bit_droit = 0
        else:
            bit_droit = 1
        return 2 * bin_to_dec(nb_bin[:-1]) + bit_droit
```
