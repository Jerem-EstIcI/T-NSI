```python
def conv_bin(n):
    '''
    >>> conv_bin(9)
    ([1,0,0,1],4)
    '''
    l=[n%2] 
    n=n//2
    while n!=0:
        l.append(n%2)
        n=n//2
    l.reverse()
    return (l,len(l))
```

```python
conv_bin(9)
```

```python
def tri_bulles(T):
    """
    >>>tri_bulles([5,7,3,9,2])
    [2,3,5,7,9]
    """
    n = len(T)
    for i in range(n-1,0,-1):
        for j in range(i):
            if T[j] > T[j+1]:
                temp = T[j]
                T[j] = T[j+1]
                T[j+1] = temp
    return T
```

```python
tri_bulles([5,7,3,9,2])
```
