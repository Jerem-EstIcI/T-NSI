## Q4

```python
def dirgelwch(a,b):
    """
    >>>dirgelwch(11,60)
    [0,11,22,33,44,55,66,6,17,28,39,50,61,1]
    """
    nb=0
    resultat=[]
    while True:
        resultat.append(nb)
        if nb==1:
            return resultat
        elif nb<b:
            nb=nb+a
        else:
            nb=nb-b
```

```python
dirgelwch(11,60)
```
