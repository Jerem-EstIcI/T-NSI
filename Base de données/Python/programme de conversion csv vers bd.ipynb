```python
# importation du fichier csv
import csv
donnees = []
with open('meilleures-audiences-en-salles-depuis-1945public.csv',encoding='utf-8') as fichiercsv:
    contenu = csv.reader(fichiercsv, delimiter=';')
    for ligne in contenu:
        donnees.append(ligne)
#Lecture du début du fichier csv
for i in donnees[0:10]:
    print(i)
```

```python
# création de la table film

# DROP TABLE <nom> -> efface la table <nom>
# CREATE TABLE <nom> -> crée une table <nom>

import sqlite3

conn = sqlite3.connect('box_office.db')

c = conn.cursor()

#si la table existe on la supprime
c.execute('DROP TABLE IF EXISTS film')

#on crée la table avec quatre attributs : trois de type entier et un de type text
#la clé primaire sera géré automatiquement
c.execute('''CREATE TABLE film (
    id_film integer primary key autoincrement,
    titre text,
    annee integer,
    entree interger)''')

#Insertion des éléments 
for ligne in range(1,len(donnees)):
    liste=[donnees[ligne][0],donnees[ligne][2],donnees[ligne][4]]
    c.execute("INSERT INTO film ('titre','annee','entree') values(?,?,?)",liste)
conn.commit()
conn.close()
```
