```python
def factorielle(n):
    resultat=1
    for i in range(2,n+1):
        resultat*=i
    return resultat
```

```python
factorielle(50)
```

```python
def factorielle(n):
    #condition d'arrêt	
    if n==1: 
        return 1
    return n*factorielle(n-1)
```

```python
factorielle(50)
```

```python
factorielle(992)
```

```python
def pgcd(a,b):
    if a < b: 
        a,b=b,a
    while b: 
        a,b=b,a%b
    return a
```

```python
def pgcd(a,b):
    #condition d'arrêt
    if a%b==0:
        return b
    return pgcd(b,a%b)
```
