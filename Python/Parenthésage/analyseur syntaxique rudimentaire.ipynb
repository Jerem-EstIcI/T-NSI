Un analyseur syntaxique rudimentaire
====================================

La fonction `validation(s)` proposée ci-dessous permet de vérifier que les diverses sortes de parenthèses `()`, `[]`,`{}` sont bien équilibrées dans la chaîne de caractères `s`.

```python
class Pile:
    def __init__(self):
        self.p=[]
    def vide(self):
        return self.p==[]
    def empile(self,x):
        self.p.append(x)
    def depile(self):
        return self.p.pop()

def validation(s):
    """
    >>> validation('')
    True
    >>> validation('({[]()})[]')
    True
    >>> validation('(({}[])')
    False
    """
    ouvrante={')':'(','}':'{',']':'['}
    p=Pile()
    for c in s:
        if c in '({[':
            p.empile(c)
        if c in ')}]':
            if p.vide():
                return False
            tos=p.depile()
            if tos!=ouvrante[c]:
                return False
    return p.vide()

import doctest
doctest.testmod()
```
