Parenthésages
=============

1. Compléter la fonction `valide` conformément à sa docstring.

```python
def valide(s):
    """
    vérifie que le parenthésage de la chaîne s est correct :
    * autant de ) que de (
    * aucune parenthèse n'est fermée avant d'être ouverte
    >>> valide('(()(()))')
    True
    >>> valide('((()()')
    False
    >>> valide('())(()')
    False
    """
    n = 0  # on part de n=0
    for caractere in s:  # pour chaque caractère dans la chaîne s
        if caractere == '(':  # si c'est une (
            n += 1  # n augmente de 1
        elif caractere == ')':  # si c'est une )
            n -= 1  # n diminue de 1
            if n < 0:  # on vérifie que n reste positif, sinon on renvoie False
                return False
    return n == 0  # on renvoie True si n est revenu à zéro, False sinon
```

```python
valide('(()(()))')
```
