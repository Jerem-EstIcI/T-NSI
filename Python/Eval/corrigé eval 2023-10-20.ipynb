**Exercice 3** Une fusion itÃ©rative

```python
def fusion(tab1,tab2):
    """
    cette fonction renvoie la fusion de deux 
    tableaux triÃ©s
    >>> fusion([14,35,50],[22,29,44,56])
    [14,22,29,35,44,50,56]
    """
    n1=len(tab1)
    n2=len(tab2)
    tab=[0]*(n1+n2)
    i1=0
    i2=0
    i=0
    while i1<n1 and i2<n2:
        if  tab1[i1]<tab2[i2]:
            tab[i]=tab1[i1]
            i=i+1
            i1=i1+1
        else:
            tab[i]=tab2[i2]
            i=i+1
            i2=i2+1
        print(n1-i1,n2-i2,n1-i1+n2-i2)
    while i1<n1:
        tab[i]=tab1[i1]
        i=i+1
        i1=i1+1
        print(n1-i1,n2-i2,n1-i1+n2-i2)
    while i2<n2 :
        tab[i]=tab2[i2]
        i= i+1
        i2= i2+1
        print(n1-i1,n2-i2,n1-i1+n2-i2)
    return  tab

fusion([14,35,50],[22,29,44,56])
```
