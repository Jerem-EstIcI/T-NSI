**Q1** DÃ©rouler la fonction dans un tableau.

````python
from tutor import tutor

def dicho(elt,tab):
    deb=0
    fin=len(tab)-1
    while deb<fin:
        mil=(deb+fin)//2
        if elt==tab[mil]:
            return True
        elif elt<tab[mil]:
            fin=mil-1
        else:
            deb=mil+1
    return False

dicho(64,[17,22,29,33,35,40,44,56,64,85])

tutor()
````

````python
import tutor

def dichotomie_rec(elt,tab):
    if tab==[]:
        return False
    m=len(tab)//2
    if elt==tab[m]:
        return True
    elif elt<tab[m]:
        return dichotomie_rec(elt,tab[:m])
    else:
        return dichotomie_rec(elt,tab[m+1:])
dichotomie_rec(56,[17,22,29,33,35,40,44,56,64,85])

tutor.tutor()
````

**Q6** Ãcrire une fonction `recherche(abr, v)` qui prend en argument un arbre binaire de recherche `abr` et un entier `v` et renvoie la valeur `True` si cet entier est une Ã©tiquette de lâarbre, `False` sinon.

````python
import tutor

racine = [18, [15, [13, [12, None, None], None], [16, None, None]], [23, [19, None, [21, None, None]], [32, None, None]]]

def recherche(abr, v):
    if not abr:
        return False

    if v == abr[0]:
        return True
    elif v < abr[0]:
        return recherche(abr[1], v)
    else:
        return recherche(abr[2], v)

result = recherche(racine, 19)
print(result)

tutor.tutor()
````

**Q7** ComplÃ©ter les **mÃ©thodes** `insere_tout`, `recherche` et `parcours_infixe` de la classe `ABR`

````python
class ABR:
    def __init__(self, valeur):
        self.data = [valeur, None, None]

    def insere(self, valeur):
        est_insere = False
        abr = self.data
        while not est_insere:
            if valeur == abr[0]:
                est_insere = True
            elif valeur < abr[0]:
                if abr[1] is not None:
                    abr = abr[1].data
                else:
                    abr[1] = ABR(valeur)
                    est_insere = True
            else:
                if abr[2] is not None:
                    abr = abr[2].data
                else:
                    abr[2] = ABR(valeur)
                    est_insere = True

    def insere_tout(self, liste):
        for e in liste:
            self.insere(e)

    def recherche(self, valeur):
        if valeur == self.data[0]:
            return True
        elif valeur < self.data[0] and self.data[1] is not None:
            return self.data[1].recherche(valeur)
        elif valeur > self.data[0] and self.data[2] is not None:
            return self.data[2].recherche(valeur)
        return False

    def parcours_infixe(self):
        resultat = []
        if self.data[1] is not None:
            gauche = self.data[1].parcours_infixe()
            resultat.extend(gauche)
        resultat.append(self.data[0])
        if self.data[2] is not None:
            droite = self.data[2].parcours_infixe()
            resultat.extend(droite)
        return resultat
````

**Q9** RÃ©Ã©crire la classe `ABR` avec le constructeur ci-dessous.

````python
class ABR:
    def __init__(self, valeur):
        self.v = valeur
        self.g = None  # Left subtree
        self.d = None  # Right subtree

    def insere(self, valeur):
        if valeur < self.v:
            if self.g is None:
                self.g = ABR(valeur)
            else:
                self.g.insere(valeur)
        elif valeur > self.v:
            if self.d is None:
                self.d = ABR(valeur)
            else:
                self.d.insere(valeur)

    def insere_tout(self, liste):
        for valeur in liste:
            self.insere(valeur)

    def parcours_infixe(self):
        resultat = []
        if self.g is not None:
            resultat.extend(self.g.parcours_infixe())
        resultat.append(self.v)
        if self.d is not None:
            resultat.extend(self.d.parcours_infixe())
        return resultat
````
