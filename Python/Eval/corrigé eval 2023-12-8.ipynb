```python
class Pile:
    def __init__(self):
        self.p=[]
    def vide(self):
        return self.p == []
    def empile(self,e):
        self.p.append(e)
    def depile(self):
        return self.p.pop()
```

```python
class ArbreBS:
    def __init__(self,etiquette,gauche=None,droite=None):
        self.e=etiquette
        self.g=gauche
        self.d=droite
    def feuille(self):
        return self.g is None and self.d is None
    def racine(self):
        return self.e
    def gauche(self):
        return self.g
    def droite(self):
        return self.d
    def valeur(self):
        if self.feuille():
            return int(self.racine())
        elif self.e=='+':
            return self.g.valeur()+self.d.valeur()
        elif self.e=='-':
            return self.g.valeur()-self.d.valeur()
        elif self.e=='*':
            return self.g.valeur()*self.d.valeur()
        elif self.e=='//':
            return self.g.valeur()//self.d.valeur()
    def ParcoursSuffixeRec(self):
        if self.feuille():
            return str(self.e)
        else:
            return self.g.ParcoursSuffixeRec()+' '+self.d.ParcoursSuffixeRec()+' '+self.e
    def valeurParcours(self):
        p=Pile()
        parcours=self.ParcoursSuffixeRec()
        t=parcours.split()
        for e in t:
            if e=='+':
                p.empile(p.depile()+p.depile())
            elif e=='-':
                x=p.depile()
                y=p.depile()
                p.empile(y-x)
            elif e=='*':
                p.empile(p.depile()*p.depile())
            elif e=='//':
                x=p.depile()
                y=p.depile()
                p.empile(y//x)
            else:
                p.empile(int(e))
        return p.depile()
```

```python
class File:
    def __init__(self):
        self.entree=Pile()
        self.sortie=Pile()
    def vide(self):
        return self.entree.vide() and self.sortie.vide()
    def enfile(self,e):
        self.entree.empile(e)
    def defile(self):
        if self.sortie.vide():
            while not self.entree.vide():
                self.sortie.empile(self.entree.depile())
        return self.sortie.depile()
```

```python
class Arbre:
    def __init__(self,racine,enfants=[]):
        self.r=racine
        self.e=enfants
    def feuille(self):
        return self.e==[]
    def racine(self):
        return self.r
    def enfants(self):
        return self.e
```

```python
def ParcoursLargeur(a):
    f=File()
    l=[]
    f.enfile(a)
    while not f.vide():
        u=f.defile()
        l.append(u.racine())
        for e in u.enfants():
            f.enfile(e)
    return(l)
```
