```python
def score_max(i,j,p):
    if i==len(p)-1:
        return p[i][j]
    return p[i][j]+max(score_max(i+1,j,p),score_max(i+1,j+1,p))
```

```python
p2=[[4],[6,2],[3,5,7],[5,1,6,2],[4,7,3,5,2]]
```

```python
def pyramide_nulle(n):
    p=[[0]]
    i=1
    while i < n:
        p.append([0 for _ in range(i+1)])
        i+=1
    return p
```

```python
def prog_dyn(p):
    n=len(p)
    s=pyramide_nulle(n)
    for j in range(n):
        s[n-1][j]=p[n-1][j]
    for i in range(n-2,-1,-1):
        for j in range(i+1):
            s[i][j]=p[i][j]+ max(s[i+1][j],s[i+1][j+1])
    return s[0][0]
        
```
