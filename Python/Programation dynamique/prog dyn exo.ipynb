```python
def score_max(i,j,p):
    if i==len(p)-1:
        return p[i][j]
    return p[i][j]+max(score_max(i+1,j,p),score_max(i+1,j+1,p))
```

```python
p2=[[4],[6,2],[3,5,7],[5,1,6,2],[4,7,3,5,2]]
```

```python
score_max(0,0,p2)
```

```python
def pyramide_nulle(n):
    p=[[0]]
    i=1
    while i < n:
        p.append([0 for _ in range(i+1)])
        i+=1
    return p
        
        
```

```python
pyramide_nulle(1)
```

```python
pyramide_nulle(5)
```
