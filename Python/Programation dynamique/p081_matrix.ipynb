```python
def chemin(m,i,j):
    '''
    prend une matrice m et renvoie la veleur du chemin le plus court pour aller de (1,1) à (i,j)
    >>>chemin(m0,5,5)
    2427
    '''
    #cas de base
    if i==0 and j==0:
        return m[0][0]
    # un appel récursif
    if i==0:
        return m[i][j]+chemin(m,i,j-1)
    if j==0:
        return m[i][j]+chemin(m,i-1,j)
    # 2 appels récursifs (i>0 j>0)
    return m[i][j]+min(chemin(m,i,j-1),chemin(m,i-1,j)) #p[i][j]+max(score_max(i+1,j,p),score_max(i+1,j+1,p))
```

```python
def n(i,j):
    if i==0 and j==0: 
        return 0
    if i>0 and j==0:
        return i
    if i==0 and j>0:
        return j
    return n(i-1,j)+n(i,j-1)+2
```
