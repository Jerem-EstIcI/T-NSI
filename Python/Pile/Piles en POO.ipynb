Piles en POO
============

L'objectif de ce notebook est d'implémenter une structure de pile en POO, puis après quelques manipulations, d'écire quelques autres méthodes.

On rappelle que les primitives du TAD pile (type abstrait de données) sont :
* la création d'une pile vide
* l'empilage d'une valeur
* le dépilage d'une valeur
* le test pour savoir si la pile est vide

On va utiliser ici une liste Python pour stocker le contenu de la pile ; le type `list` possède en fait déjà toutes les méthodes pour pouvoir se comporter comme une pile.

On examinera ultérieurement d'autres façons de représenter une pile.

```python
class Pile:
    def __init__(self):
        self.p=[]
    def vide(self):
        return self.p==[]
    def empiler(self,e):
        self.p.append(e)
    def depiler(self):
        return self.p.pop()
    def __repr__(self):
        return ' : '.join(str(k) for k in self.p)+' <-- TOS'
    def __len__(self):
        return len(self.p)
    def dup(self):
        tos=self.depiler()
        self.empiler(tos)
        self.empiler(tos)
    def swap(self):
        a=self.depiler()
        b=self.depiler()
        self.empiler(a)
        self.empiler(b)
    def relegation(self):
        tos=self.depiler()
        pile2=Pile()
        while not self.vide():
            pile2.empiler(self.depiler())
        self.empiler(tos)
        while not pile2.vide():
            self.empiler(pile2.depiler())
```
