```python
def recherche(elt,abr):
    if elt==racine(abr):
        return True
    if elt<racine(abr):
        gauche=sabrg(abr)
        if gauche:
            return recherche(elt,gauche)
    else:
        droit=sabrd(abr)
        if droit:
            return recherche(elt,droit)
    return False
```

```python
def taille(abr):
    t=1
    gauche = sabrg(abr)
    droite = sabrd(abr)
    if gauche:
        t=t+taille(gauche)
    if droite:
        t=t+taille(droite)
    return t 
```

```python
def hauteur(abr):
    gauche = sabrg(abr)
    droite = sabrd(abr)
    if gauche:
        g=hauteur(gauche)
    else:
        g=-1
    if droite:
        d=hauteur(droite)
    else:
        d=-1
    return 1+max(g,d)
    
```

```python
def maxabr(abr):
    m=racine(abr)
    while sabrd(abr):
        return maxabr(sabrd(abr))
    return m
```
