```python
class Graphe:
    def __init__(self,sommets):
        self.v=[s for s in sommets]
        self.e={s:[] for s in sommets}
    def aretes(self,aretes):
        for a in aretes:
            self.e[a[0]].append(a[1])
            self.e[a[1]].append(a[0])
    def sommets(self):
        return self.v
    def voisins(self,s):
        return self.e[s]

class File:
    def __init__(self):
        self.e=[]
        self.s=[]
    def vide(self):
        return self.e==[] and self.s==[]
    def enfile(self,x):
        self.e.append(x)
    def defile(self):
        if self.s==[]:
            while self.e!=[]:
                self.s.append(self.e.pop())
        return self.s.pop()
    
def cycle(g):
    """
    fonction booléenne qui détecte un cycle dans un graphe g
    >>> g=Graphe('ABCDEFGHI')
    >>> g.aretes(['AB','AD','AG','AH','BC','BH','CD','DE','DH','DI','EF','EH','EI','FG','FI','GH','HI'])
    >>> cycle(g)
    True
    >>> g2=Graphe('0123456789')
    >>> g2.aretes(['01','12','13','04','45','46','07','78','79'])
    >>> cycle(g2)
    False
    """
    # créer une file vide f
    f=File()
    # créer un liste vide l
    l=[]
    # créer un ensemble vide dejavu
    dejavu=set()
    # enfiler un sommet de départ dans la file
    s=g.sommets()[0]
    f.enfile(s)
    # ajouter le sommet de départ à l'ensemble
    dejavu.add(s)
    # tant que la file n'est pas vide
    while not f.vide():
        # défiler un sommet u
        u=f.defile()
        # ajouter u à la fin de la liste
        l.append(u)
        # pour chaque voisin v du sommet
        for v in g.voisins(u):
            # si v n'est pas dans dejavu
            if v not in dejavu:
                # enfiler v dans f
                f.enfile(v)
                # ajouter v à dejavu
                dejavu.add(v)
            # sinon si v n'est pas dans l
            elif v not in l : 
                # renvoyer Vrai
                return True
    # renvoyer Faux
    return False
```
