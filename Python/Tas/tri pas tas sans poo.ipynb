Tri par tas avec un tas maximal sans POO
====================================================

* On va commencer par écrire deux fonctions pour heapifier un tableau ;
* ensuite on va écrire un tri par tas en place.

Attention, on travaille à partir de l'indice zéro donc le parent d'un sommet numéroté `k` sera le sommet numéroté `(k+1)//2` et ses enfants seront numérotés `2*k+1` et `2*k+2`.

On va commencer par représenter la liste `[80,76,14,50,35,22,29,56,44,85]` sous forme d'un arbre complet dont la liste donne un parcours en largeur. Ce sera notre cas d'usage pour la fonction `heapify`.

```python
def sink(t,k,n=None):
    """
    cette fonction prend un tableau et transforme l'arbre de racine t[k] en tas,
    en supposant que les deux sous-arbres ont déjà la propriété de tas
    >>> t=[80,76,14,50,35,22,29,56,44,85]
    >>> sink(t,4)
    >>> t
    [80, 76, 14, 50, 85, 22, 29, 56, 44, 35]
    """
    if n is None:
        n=len(t)
    # si le noeud k n'a pas d'enfants
    if n<=2*k+1:
        # on a fini
        return
    # si le noeud k a un enfant gauche seulement
    if n==2*k+2:
        # si le noeud k a une étiquette plus petite que celle de l'enfant gauche
        if t[k]<t[2*k+1]:
            # échanger les deux étiquettes
            t[k],t[2*k+1]=t[2*k+1],t[k]
            print(f"sink {k}<->{2*k+1} : {t}")
        # on a fini
        return
    # sinon il y a deux enfants
    # si l'étiquette de k est plus grande que les deux étiquettes des enfants
    if t[k]>t[2*k+1] and t[k]>t[2*k+2]:
        # on a fini
        return
    # soit l le numéro de l'enfant avec la plus grande étiquette
    if t[2*k+1]>t[2*k+2]:
        l=2*k+1
    else:
        l=2*k+2
    # échanger t[k] et t[l]
    t[k],t[l]=t[l],t[k]
    print(f"sink {k}<->{l} : {t}")
    # appeler récursivement sink(t,l,n)
    sink(t,l,n)
```

```python
def heapify(t):
    for k in range(len(t)//2,-1,-1):
        sink(t,k)
    return t
```

Exécuter `heapify` sur le cas d'usage `[80,76,14,50,35,22,29,56,44,85]` **à la main**.

```python
heapify([80,76,14,50,35,22,29,56,44,85])
```

```python
def tri(t):
    print('----------début heapify-----------------')
    heapify(t)
    print('----------fin heapify-------------------')
    k=len(t)-1
    while k>0:
        t[0],t[k]=t[k],t[0]
        print(f"extraction de {t[k]} :  {t}")
        sink(t,0,k)
        k=k-1
    return t
```

Exécuter cet algorithme sur la liste `[80,76,14,50,35,22,29,56,44,85]` **à la main**.

```python
tri([80,76,14,50,35,22,29,56,44,85])
```
