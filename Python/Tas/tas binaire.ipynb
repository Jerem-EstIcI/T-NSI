Tas binaires
==========

Les feuilles sont les sommets qui n'ont pas d'enfant gauche (et donc pas d'enfant droit, soit pas d'enfant du tout). On les caractérise par leur **indice** `i` qui vérifie `2*i+1>=len(t)`, ou leur **position** qui vérifie `2*pos>=len(t)`. Par conséquent leur indice vérifie `i>=len(t)//2` et leur position `pos>len(t)//2`.
Il y a `len(t)//2` noeuds intérieurs, numérotés par les indices de `0` à `len(t)//2-1` (et les positions de `1` à `len(t)//2`), on pourra donc parcourir ces noeuds intérieurs avec un `range(len(t)//2)`.


```python
def verifie(t):
    # pour chaque noeud intérieur
    for i in range(len(t)//2):
        # si ce noeud a un fils gauche et que celui-ci lui est inférieur
        if t[2*i+1]<t[i]:
            # renvoyer faux
            return False
        # si ce noeud intérieur a un fils droit et que celui-ci lui est inférieur
        if 2*i+2<len(t) and t[2*i+2]<t[i]:
            # renvoyer faux
            return False
    # renvoyer vrai
    return True
```

```python
verifie([44,22,76,14,29,50,80,56,85])
```

```python
verifie([6,13,11,30,83,66])
```

```python
def sink(t,i):
    while i<=len(t)//2:
        j=i
        if t[2*i-1]<t[i-1]:
            j=2*i
        if 2*i<len(t) and t[2*i]<t[j-1]:
            j=2*i+1
        if j==i:
            break
        else:
            t[i-1],t[j-1]=t[j-1],t[i-1]
            i=j
```

```python
def heapify(t):
    for i in range(len(t)//2-1,0,-1):
        sink(t,i)
        print(t)
```
