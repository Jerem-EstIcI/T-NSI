```python
def verification(texte,motif,position):
    """
    >>>verification('abracadabra','ra',5)
    False
    >>>verification('abracadabra','ra',2)
    True
    """
    i=0
    while i<len(motif):
        if texte[position+i]!=motif[i]:
            return False
        else:
            i=i+1
    return True
    
```

```python
verification('abracadabra','ra',5)
```

```python
verification('abracadabra','ra',2)
```
