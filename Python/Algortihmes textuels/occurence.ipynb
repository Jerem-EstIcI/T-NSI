```python
def verification(texte,motif,position):
    """
    fonction booléenne qui renvoie True si le motif est présent dans le texte à la position
    >>>verification('abracadabra','ra',5)
    False
    >>>verification('abracadabra','ra',2)
    True
    """
    for i,lettre in enumerate(motif):
        if texte[position+i] != lettre:
            return False
    return True
```

```python
verification('abracadabra','ra',5)
```

```python
verification('abracadabra','ra',2)
```

```python
def recherche(texte,motif):
    '''
    foctione booléenne qui teste la présence de motif dans texte
    >>>recherche('abracadabra','rata')
    False
    >>>recherche('abracadabra','bra')
    True
    '''
    for position in range(len(texte)-len(motif)+1):
        if verification(texte,motif,position):
            return True
    return False
```

```python
recherche('abracadabra','ra')
```

```python
recherche('abracadabra','rata')
```

```python
recherche('abracadabra','bra')
```
