```python
def recherche(elt,abr):
    if elt==racine(abr):
        return True
    if elt<racine(abr):
        gauche=sabrg(abr)
        if gauche:
            return recherche(elt,gauche)
    else:
        droit=sabrd(abr)
        if droit:
            return recherche(elt,droit)
    return False
```

```python
def taille(abr):
    t=1
    gauche = sabrg(abr)
    droite = sabrd(abr)
    if gauche:
        t=t+taille(gauche)
    if droite:
        t=t+taille(droite)
    return t 
```

```python
def hauteur(abr):
    gauche = sabrg(abr)
    droite = sabrd(abr)
    if gauche:
        g=hauteur(gauche)
    else:
        g=-1
    if droite:
        d=hauteur(droite)
    else:
        d=-1
    return 1+max(g,d)
    
```

```python
def construit(tag,sabrg=None,sabrd=None): #paramètres par défauts

    return[tag,sabrg,sabrd]
def racine(abr): 
    return abr[0] 
def sabrg(abr): 
    return abr[1] 
def sabrd(abr): 
    return abr[2]
```

```python
def infixe(abr):
    gauche = sabrg(abr)
    droite = sabrd(abr)
    if gauche:
        g=infixe(gauche)
    else:
        g=[]
    if droite:
        d=infixe(droite)
    else:
        d=[]
    return g+[racine(abr)]+d
```
